```python
// solution 1
object Solution {
    def main(args: Array[String]) {
        val Array(n,k)=readLine.split(" ").map(_.toInt)
        val c=readLine.split(" ").map(_.toInt).sorted.reverse
        
        var total=0
        for(i<-0 until n){
            total+=(i/k+1)*c(i)
        }
        println(total)
    }
}
​
// solution 2
object Solution {
  def main(args: Array[String]) {
    import java.util.Scanner
    val sc = new Scanner(System.in)
    val (n, k) = (sc.nextInt, sc.nextInt)
    val ci = (for (i <- 0 until n) yield BigInt(sc.nextInt))
    var x = 0
    println(ci.sortWith(_ > _).grouped(k).foldLeft(BigInt(0))((a, b) => {x += 1; a + b.sum * x}))
  }
}
​
​
// solution 3
object Solution {
​
​
  def main(args: Array[String]) {
    val iter = io.Source.stdin.getLines
    val params = iter.next.split(" ")
    val num_flowers = params(0).toInt
    val num_friends = params(1).toInt
    val cost = iter.next.split(" ").map(x => x.toInt).sorted.reverse
    var sum = 0
    for (i <- 0 until cost.length) sum += cost(i) * ((i / num_friends) + 1)
    println(sum)
  }
}
​
​
// solution 4
object Solution extends App {
  val firstLine = readLine.split(" ").map(_.toInt)
  val (flowers, buyers) = (firstLine(0), firstLine(1))
  val costs = readLine.split(" ").map(_.toInt).sorted
​
  var factor = 1
  var sum = 0
  for (i <- flowers - 1 to 0 by -1) {
    sum += factor * costs(i)
    if ((flowers - i) % buyers == 0) factor += 1
  }
  println(sum)
}
​
​
// solution 5
object Solution {
  def main(args: Array[String]) {
    val Array(n, k) = readLine.split(" ").map(_.toInt)
    var c = readLine.split(" ").map(_.toInt).sortWith(_ > _)
    println((for (i <- 0 to n - 1) yield c(i) * ((i / k) + 1)).sum) 
  }
}
​
​
// solution 6
import scala.io.Source
​
object Solution {
    val DEBUG = false
    val LCOAL_TEST = false
    val FILE = "flower"
    
    var scan: Scanner = _
    
    // *********************
    // Helper field above
    // *********************
​
    def main(args: Array[String]) {
        var source = 
            if (LCOAL_TEST) {
                Source.fromFile(FILE)
            } else  {
                Source.stdin
            }
        scan = new Scanner(source)
        
        solver
    }
    
    def solver() {
        val N = scan.nextInt
        val K = scan.nextInt
        
        var prices = List[Int]()
        for (i <- 1 to N) {
            prices = scan.nextInt :: prices
        }
        
        prices = prices.sorted
        val sortedP = prices.reverse.toArray
        
        var cost: Long = 0
        for (i <- 0 until sortedP.length) {
            cost += sortedP(i) * (i / K + 1)
        }
        
        println(cost)
    }
    
    
    //*************************************
    // Common Helper methods
    //*************************************
    def multiCaseSolver() {
        for (i <- 1 to scan.nextInt ) {
            solver
        }
    }
    
    def debug(args: Any*) {
        if (DEBUG) {
            for (a <- args) {
                print(a + " ")
            }
            println()
        }
    }
    
    class Scanner(source: Source) {
        var sourceIt = source.getLines
        var pointer = getNextLine
        
        def nextToken: String = {
            if (! pointer.hasNext) pointer = getNextLine;
            return pointer.next
        }
        
        def nextInt: Int = {
            return nextToken.toInt
        }
        
        def nextLong: Long = {
            return nextToken.toLong
        }
        
        private def getNextLine(): Iterator[String] = {
            return sourceIt.next().split(" ").toIterator
        }
        
    }
    
}
```
