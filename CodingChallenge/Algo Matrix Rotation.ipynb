```python
# Solution 1
def iterate_circle(mat, m, n, i):
    for j in range(i, m - i - 1):   yield (j, i)
    for j in range(i, n - i - 1):   yield (m - i - 1, j)
    for j in range(m - i - 1, i, -1):   yield (j, n - i - 1)
    for j in range(n - i - 1, i, -1):   yield (i, j)
​
def rotate(mat, m, n, r):
    for i in range(min(m, n) / 2):
        circle_length = 2 * (m + n - 4 * i) - 4
        offset = circle_length - (r % circle_length)
        if offset == circle_length:
            continue
        line = [mat[x][y] for (x, y) in iterate_circle(mat, m, n, i)]
        for (x, y) in iterate_circle(mat, m, n, i):
            mat[x][y] = line[offset]
            offset = (offset + 1) % circle_length
​
m, n, r = map(int, raw_input().split())
mat = [[int(j) for j in raw_input().split()] for i in range(m)]
rotate(mat, m, n, r)
for i in xrange(m):
    for j in xrange(n):
        print mat[i][j],
    print
​
# Solution 2
from __future__ import print_function
​
M, N, R = map(int, raw_input().split())
​
mat = []
out = []
​
# Build input and output arrays
for row in range(M):
    mat.append(map(int, raw_input().split()))
    out.append(list([0 for _ in range(N)]))
​
def new_index(r, c):
    """Returns a pair `(r', c')`, the new indices of element `a_rc`.
    Has O(1) space and time complexity.
    """
    ring_num = min(r, c, (M - 1) - r, (N - 1) - c)
    R_rc = R % (2*(M - 2*ring_num) + 2*(N - 2*ring_num) - 4)
​
    # Rotation functions for each side. Each function shifts
    # r, c by at most R while staying on the same side, returning
    # a pair (row delta, column delta)
    rot_funcs = [
        lambda r, c, R: (0, -min(R, c - ring_num)),
        lambda r, c, R: (min(R, (M - 1) - ring_num - r), 0),
        lambda r, c, R: (0, min(R, (N - 1) - ring_num - c)),
        lambda r, c, R: (-min(R, r - ring_num), 0),
    ]
​
    # Finds the new index corresponding to (r, c) by shifting along
    # the four sides of the ring corresponding to (r, c).
    # Runs in O(1).
    def shift_along_sides(f_i, R, r, c):
        # Note that since R_rc is at most the perimeter of the "ring",
        # This has at most 4 iterations and therefore runs in O(1).
        while R > 0:
            r_delta, c_delta = rot_funcs[f_i](r, c, R)
            R = R - abs(r_delta) - abs(c_delta)
            r, c = r + r_delta, c + c_delta
            f_i = (f_i + 1) % len(rot_funcs)
        return r, c
​
    # Compute starting side (as an index into rot_funcs)
    if r == ring_num:
        f_i = 0
    elif c == ring_num:
        f_i = 1
    elif r == M - ring_num - 1:
        f_i = 2
    else:
        f_i = 3
​
    return shift_along_sides(f_i, R_rc, r, c)
​
for row in range(M):
    for col in range(N):
        r, c = new_index(row, col)
        out[r][c] = mat[row][col]
​
for row in range(M):
    for col in range(N):
        print(out[row][col], end="")
        if col < N - 1:
            print(" ", end="")
    print()
```
