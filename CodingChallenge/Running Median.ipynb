```python
# solution 1 (concise, clear, using 2 heapqs)
from heapq import heappop, heappush, heappushpop
            
N = int(input())
S = [int(input()) for i in xrange(N)]
L, R  = [], [S[0]]
​
for i in xrange(N):
    print "%.1f" % R[0] if len(R)>len(L) else (R[0]-L[0])/2.0
    if i==N-1: break
    heappush(L,-heappushpop(R, S[i+1]))
    if len(L)>len(R): heappush(R,-heappop(L))
​
​
​
​
​
​
#solution 2 (more verbose, same idea)
import heapq
​
minheap = []
maxheap = []
​
N = int(input())
​
# minheap has all elements greater than the median
# maxheap has all elements less than the median
# min, max
# [], [] incoming is 4
# [4], [], read 4, incoming is 7
# [4], [7], read avg, incoming is 5
# [5, 4] [7], incoming is 6
# [5, 4] [6, 7]
# Algorithm:
# Find which list has the median
# If incoming is less than median put in max, else put in min
# if one differs by more than 1 from other, rebalance elements
​
for _ in xrange(N):
    x = int(input())
    def get_median():
        if len(minheap) > len(maxheap):
            return minheap[0] * 1.0
        elif len(minheap) < len(maxheap):
            return -maxheap[0] * 1.0
        else:
            if not minheap or not maxheap:
                return 0
            return (minheap[0] - maxheap[0]) / 2.0
    median = get_median()
    if x < median:
        heapq.heappush(maxheap, -x)
    else:
        heapq.heappush(minheap, x)
        
    if(len(maxheap) - len(minheap) == 2):
        heapq.heappush(minheap, -heapq.heappop(maxheap))
    elif (len(minheap) - len(maxheap) == 2):
        heapq.heappush(maxheap, -heapq.heappop(minheap))
    print get_median()
​
​
​
​
​
​
# solution 3 (using bisect)
import bisect
​
N = int(raw_input())
​
l = []
​
for i in xrange(N):
    bisect.insort(l, int(raw_input()))
    l_len = len(l)
    middle = l_len / 2
    elmt = l[middle] + 0.0
    
    if l_len % 2 == 0:
        elmt += l[middle-1]
        elmt /= 2.0
    
    elmt_s = str(elmt)
    p_ind = elmt_s.index('.')
    print elmt_s[:p_ind+2]
​
​
​
​
​
​
​
​
# solution 4 (simple median class)
class Median(object):
    def __init__(self):
        self.num = 0
        self.even = True
        self.sortedList = []
    
    def insert(self, val):
        low , high = 0, self.num
        mid = 0
        while low<high:
            mid = (low+high) /2
            if val >= self.sortedList[mid]:
                low = mid + 1
            else:
                high = mid
        self.sortedList.insert(low, val)
        self.num += 1
        if (self.even):
            self.even = False
            median_index = self.num/2
            return self.sortedList[median_index]
        else:
            self.even = True
            median_index = int(self.num/2)
            return (self.sortedList[median_index]+self.sortedList[median_index-1])/2.0
        
if __name__ == '__main__':
    num = input()
    median = Median()
    for i in xrange(num):
        val = float(input())
        print median.insert(val)
        
​
​
​
​
# solution 5 (defining heap class from scratch)
class MinHeap:
    def __init__(self):
        self.heapList = [-1000000000]
        self.currentSize = 0
        
    def percUp(self,i):
        while self.heapList[i] < self.heapList[i // 2]:
            tmp = self.heapList[i // 2]
            self.heapList[i // 2] = self.heapList[i]
            self.heapList[i] = tmp
            i = i // 2        
    def insert(self,k):
        self.heapList.append(k)
        self.currentSize = self.currentSize + 1
        self.percUp(self.currentSize)        
    def percDown(self,i):
        while (i * 2) <= self.currentSize:
            mc = self.minChild(i)
            if self.heapList[i] > self.heapList[mc]:
                tmp = self.heapList[i]
                self.heapList[i] = self.heapList[mc]
                self.heapList[mc] = tmp
            i = mc
    def minChild(self,i):
        if i * 2 + 1 > self.currentSize:
            return i * 2
        else:
            if self.heapList[i*2] < self.heapList[i*2+1]:
                return i * 2
            else:
                return i * 2 + 1
    def delMin(self):
        retval = self.heapList[1]
        self.heapList[1] = self.heapList[self.currentSize]
        self.currentSize = self.currentSize - 1
        self.heapList.pop()
        self.percDown(1)
        return retval
    def getRoot(self):  
        return self.heapList[1]    
​
​
class MaxHeap:
    def __init__(self):
        self.heapList = [1000000000]
        self.currentSize = 0
​
    def percUp(self,i):
         while i // 2 > 0:
            if self.heapList[i] > self.heapList[i // 2]:
                tmp = self.heapList[i // 2]
                self.heapList[i // 2] = self.heapList[i]
                self.heapList[i] = tmp
            i = i // 2
    def insert(self,k):
        self.heapList.append(k)
        self.currentSize = self.currentSize + 1
        self.percUp(self.currentSize)
    def percDown(self,i):
         while (i * 2) <= self.currentSize:
            mc = self.maxChild(i)
            if self.heapList[i] < self.heapList[mc]:
                tmp = self.heapList[i]
                self.heapList[i] = self.heapList[mc]
                self.heapList[mc] = tmp
            i = mc
    def maxChild(self,i):
         if i * 2 + 1 > self.currentSize:
            return i * 2
         else:
            if self.heapList[i*2] > self.heapList[i*2+1]:
                return i * 2
            else:
                return i * 2 + 1
    def delMax(self):
        retval = self.heapList[1]
        self.heapList[1] = self.heapList[self.currentSize]
        self.currentSize = self.currentSize - 1
        self.heapList.pop()
        self.percDown(1)
        return retval   
    def getRoot(self):  
        return self.heapList[1]       
            
                   
N = int(raw_input())
a = float(raw_input())
print "%.1f" % a
b = float(raw_input())
res= (a+b)/2
print "%.1f" % res
myMinHeap = MinHeap() # The larger numbers
myMaxHeap = MaxHeap() # The smaller numbers
myMinHeap.insert(max(a,b)) # The larger numbers
myMaxHeap.insert(min(a,b)) # The smaller numbers
for i in range(2,N):
​
    toInsert = float(raw_input())
    if toInsert < myMaxHeap.getRoot():
        myMaxHeap.insert(toInsert)
    else:
        myMinHeap.insert(toInsert)
​
    if myMaxHeap.currentSize>myMinHeap.currentSize+1:
        myMinHeap.insert(myMaxHeap.delMax())
    elif myMinHeap.currentSize>myMaxHeap.currentSize+1:
        myMaxHeap.insert(myMinHeap.delMin())
​
    if myMaxHeap.currentSize>myMinHeap.currentSize:
        print "%.1f" % myMaxHeap.getRoot()
    elif myMinHeap.currentSize>myMaxHeap.currentSize:
        print "%.1f" % myMinHeap.getRoot()
    else:
        res = (myMaxHeap.getRoot() +myMinHeap.getRoot() )/2
        print "%.1f" % res
```
