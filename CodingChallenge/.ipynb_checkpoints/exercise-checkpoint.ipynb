#fibunaci

```python
def fib(n):
    if n < 0 : return "Error"
    elif n == 0 : return 0
    elif n == 1 : return 1
    else : return fib(n-1)+fib(n-2)
```

#Binary Tree

```python
class BinaryTreeNode:

    def __init__(self, value):
        self.value = value
        self.left  = None
        self.right = None

    def insert_left(self, value):
        self.left = BinaryTreeNode(value)
        return self.left

    def insert_right(self, value):
        self.right = BinaryTreeNode(value)
        return self.right
```

```python
node = BinaryTreeNode(5)
```

```python
node.insert_left(2)
```

```python
node.insert_right(3)
```
