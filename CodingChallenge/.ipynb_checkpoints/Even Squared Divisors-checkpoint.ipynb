```python
# Editorial
<h3>Smart Strategy</h3>
​
<p><br>
We can reduce the previous strategy by precomputing the number of divisors and even perfect square divisors for every number by using a idea called a <a href="http://en.wikipedia.org/wiki/Sieve_of_Eratosthenes">sieve</a>. We consider each number in turn, and mark its multiples. <br>
<br>
This sieve is implemented in the following code:  </p>
​
<pre><code>from math import sqrt
from fractions import gcd
​
N = 1000000
​
# calculate even squares
is_even_square = [False]*(N+1)
for i in xrange(1,int(sqrt(N))+1):
    if i % 2 == 0:
        is_even_square[i * i] = True
​
# calculate even square divisors of numbers using a sieve, in O(N log N)
even_square_divisors = [0]*(N+1)
divisors = [0]*(N+1)
for i in xrange(1,N+1):
    for j in xrange(i&lt;&lt;1,N+1,i):
        divisors[j] += 1
    if is_even_square[i]:
        for j in xrange(i&lt;&lt;1,N+1,i):
            even_square_divisors[j] += 1
​
# calculate answers
for cas in xrange(input()):
    n = input()
    if even_square_divisors[n] == 0:
        print 0
    else:
        p = even_square_divisors[n]
        q = divisors[n]
        g = gcd(p, q)
        print "%s/%s" % (p / g, q / g)
</code></pre>
​
# Editorial 2
from math import sqrt
    from fractions import gcd
​
    N = 1000000
​
    # calculate even square divisors of numbers using a sieve, in O(N log log N)
    even_square_divisors = [0]*(N+1)
    divisors = [1]*(N+1)
​
    # initialize
    for i in xrange(4,N+1,4):
        even_square_divisors[i] = 1
​
    # special loop for 2
    pow2 = 2
    e = 1
    while pow2 <= N:
        for j in xrange(pow2,N+1,pow2):
            divisors[j] += divisors[j] / e
            if e >= 4 and e % 2 == 0:
                even_square_divisors[j] += even_square_divisors[j]/(e/2-1)
        pow2 *= 2
        e += 1
​
    # other primes
    for i in xrange(3,N+1):
        if divisors[i] == 1:
            # i is prime
            powi = i
            e = 1
            while powi <= N:
                for j in xrange(powi,N+1,powi):
                    divisors[j] += divisors[j] / e
                    if e % 2 == 0:
                        even_square_divisors[j] += even_square_divisors[j]/(e/2)
                powi *= i
                e += 1
​
        divisors[i] -= 1
​
    # adjust for actual even squares
    for i in xrange(2,int(sqrt(N))+1,2):
        even_square_divisors[i * i] -= 1
​
    # calculate answers
    for cas in xrange(input()):
        n = input()
        if even_square_divisors[n] == 0:
            print 0
        else:
            p = even_square_divisors[n]
            q = divisors[n]
            g = gcd(p, q)
            print "%s/%s" % (p / g, q / g)
​
# solution 1
import sys
import fractions
from math import ceil
from math import sqrt
​
def parseInt(stream):
    return int(stream.readline().rstrip())
​
def genFactors(n):    
    return set(reduce(list.__add__, 
                [ [i, n//i] for i in range(1, int(n**0.5) + 1) if n % i == 0 ] ) )
​
def isEvenPerfectSquare(n):
    return n % 2 == 0 and ceil(sqrt(n)) == sqrt(n)
​
if __name__ == "__main__":
    stream = sys.stdin
    t = parseInt(stream)
    for i in range(t):
        n = parseInt(stream)
        f = genFactors(n)
        f.remove(n)
        reqDiv = 0
        allDiv = len(f)
        for i in f:
            if(isEvenPerfectSquare(i)):
                reqDiv += 1
        
        prob = fractions.Fraction(reqDiv,allDiv)
        prob = str(prob)
        print prob
​
​
​
# solution 2
from fractions import *
​
issquare = lambda n: int(n**.5) == n**.5
n = 1000001
​
d = [1]*(n+1)
e = [0]*(n+1)
for divisor in xrange(2,n+1):
  if issquare(divisor) and divisor%2==0:
    for multiple in xrange(divisor*2,n+1,divisor):
      e[multiple] += 1
      
  for multiple in xrange(divisor*2,n+1,divisor):
    d[multiple] += 1
    
T = int(raw_input())
for i in xrange(T):
    n = int(raw_input())
    if e[n] == 0:
        print 0
    else:
        print str(Fraction(e[n],d[n]))
​
​
    
      
            
​
​
# solution 3
import sys
import math
from fractions import Fraction
data=sys.stdin.readlines()
T=int(data[0])
​
def is_square(apositiveint):
    x = apositiveint // 2
    seen = set([x])
    while x * x != apositiveint:
        x = (x + (apositiveint // x)) // 2
        if x in seen: return False
        seen.add(x)
    return True
​
def divisors(n):
    res = []
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            res.append(n//i)
            res.append(i)
    return set(res)
​
def fun(p):
    num = [is_square(i) for i in p if i > 2 and i%2==0]
    x = num.count(True)
    y = len(k)+1
    print Fraction(x,y)
    
        
for i in range(1,T+1):
    k=divisors(int(data[i].strip()))
    #print k
    fun(k)   
    
    
# solution 4
import fractions
​
evenSquares = set([(2*n)**2 for n in range(501)])
​
for _ in range(int(raw_input())):
    N = int(raw_input())
    nFactors = 1
    evenSquareFactors = 0
    for factor in range(2, int(N ** .5) + 1):
        if N % factor == 0:
            nFactors += 1
            evenSquareFactors += (factor in evenSquares)
            if N // factor != factor:
                nFactors += 1
                evenSquareFactors += ((N//factor) in evenSquares)
    f = fractions.Fraction(evenSquareFactors, nFactors)
    print f
​
​
​
​
​
​
​
# solution 5
import sys
import fractions
from math import ceil
from math import sqrt
​
# Parsing functions 
'''
    Reads in an integer from stream passed in as the parameter. Simple
    parsing function that can read from files as well as standard input
''' 
def parseInt(stream):
    return int(stream.readline().rstrip())
​
'''
    Generates factors of a particular number by iterating through the num
    upto sqrt(num) . Efficient approach using prime factor generation. The
    function reduce if new to you, just computes result of the supplied
    function f (in this case list.__add__) on the first two elements of the
    supplied list, obtains a result and then performs the same function on
    the result and the next element in the list. In this way it reduces the 
    entire list to one result
'''
def genFactors(n):    
    return set(reduce(list.__add__, 
                [ [i, n//i] for i in range(1, int(n**0.5) + 1) if n % i == 0 ] ) )
'''
    Core Function. Checks if a number is both even as well as a perfect 
    square. Returns true if yes, False other wise via a boolean comparison
'''
def isEvenPerfectSquare(n):
    return n % 2 == 0 and ceil(sqrt(n)) == sqrt(n)
​
# Main function for the program
if __name__ == "__main__":
    stream = sys.stdin
    t = parseInt(stream)
    for i in range(t):
        n = parseInt(stream)
        f = genFactors(n)
        f.remove(n)
        reqDiv = 0
        allDiv = len(f)
        for i in f:
            if(isEvenPerfectSquare(i)):
                reqDiv += 1
        
        prob = fractions.Fraction(reqDiv,allDiv)
        prob = str(prob)
        print prob
​
​
​
​
# solution 6
import sys, math
​
def euclid(a, b):
    if b == 0:
        return a
    return euclid(b, a % b)
​
def divisors(n):
    ret = set()
    for x in range(1, int(math.sqrt(n)) + 1):
        if n % x == 0:
            ret.add(x)
            ret.add(n / x)
​
    ret.remove(n)
    return sorted(list(ret))
    
​
def main():
    t = int(sys.stdin.readline())
    for _ in range(t):
        n = int(sys.stdin.readline())
        divlist = divisors(n)
        x = 0
        y = len(divlist)
        for i in divlist:
            if math.sqrt(i) == int(math.sqrt(i)) and i % 2 == 0:
                x += 1
        div = euclid(x, y)
        if x == 0:
            print 0
        else:
            print (str(x / div) + '/' + str(y / div))
        
if __name__ == '__main__':
    main()
​
​
​
​
​
​
​
# solution 7
from fractions import Fraction
def nof(n):    
    return len(set(reduce(list.__add__, 
            ([i, n//i] for i in range(1, int(n**0.5) + 1) if n % i == 0))))
li=[]
i=2
f=1
while f<10**6:
    f=i**2
    li.append(f)
    i+=2
for _ in xrange(input()):
    n=input()
    i=0
    ct=0
    while li[i]<=n/2:
        if n%li[i]==0:
            ct+=1
        i+=1
    if ct==0:
        print 0
    else:
        print Fraction(ct,nof(n)-1)
​
​
​
# solution 8
from math import sqrt
from collections import defaultdict
from fractions import Fraction
​
def sieve_primes(N):
    is_prime = [True for x in range(N+1)]
    is_prime[0] = is_prime[1] = False
    for p in range(2,int(sqrt(N))+2):
        if not is_prime[p]: continue
        for x in range(p*p,N+1, p):
            is_prime[x] = False
    return [p for p in range(N+1) if is_prime[p]]
​
def prime_factorization(n, primes):
    pfact = defaultdict(int)
    for p in primes:
        while n%p == 0:
            pfact[p] += 1
            n /= p
        if p > sqrt(n):
            break
    if n > 1:
        pfact[n] += 1
    return pfact
​
def is_square(n):
    r = int(sqrt(n))
    return r*r == n or (r+1)*(r+1) == n
​
primes = sieve_primes(1000)
t = int(raw_input())
for _ in range(t):
    n = int(raw_input())
    if n%4 != 0:
        print 0
        continue
    pfact = prime_factorization(n,primes)
    num_div, num_even_square = 1,1
    for p in pfact.keys():
        if p == 2: num_even_square *= pfact[2]//2
        else: num_even_square *= pfact[p]//2 + 1
        num_div *= pfact[p] + 1
    if is_square(n) and n % 2 == 0: num_even_square -= 1
    if num_even_square == 0: print 0
    else: print Fraction(num_even_square, num_div-1)
```
