```python
# solution 1
from math import factorial
​
def scores(A, B):
    if A == B or (A < 25 and B < 25) or (abs(A-B)!= 2 and A >= 23 and B >= 23):
        return 0
    
    if A < B:
        A, B = B, A
    
    ## total number of ways to 24:B
    ## permuations of a string with 24 "A"s and B "B"s then we put "A" at the end
    ## to denote the sequence
    p = 10**9 + 7
    if A == 25:
        return factorial(24 + B)/ factorial(24)/ factorial(B) % p
    
    ## total number of ways up to 24:24
    total_to_24_all = factorial(48)/factorial(24)**2 % p
    
    ## total number of ways from 24:24 to final_score
    ## we must always get to 25:25, 26:26 .., x:x, x+1:x, x+2:x (1 way)
    ## There are 2 choices for each 25:24, 26:25, .. x-1:x (A-26 ways)
    total_to_final_from_24_all = pow(2, (A-26), p)
    
    return total_to_24_all * total_to_final_from_24_all % p
​
A = input()
B = input()
print scores(A, B)
​
# solution 2
def volleyball(a, b):
    # call the winner a, without losing information
    if a<b: a,b = b,a
    # regular win 
    if a == 25 and b <24: 
        return combo(a-1, b) % (10**9+7)
    # win by tie breaker from 24:24
    if a > 25 and a-b == 2:
        return (combo(24, 24) * 2**(a-26)) % (10**9+7)
    # not a valid final score
    return 0
    
m = {}
def combo(a,b):
    # either way is the same number
    if a<b: a,b = b,a
    # trivial 
    if b == 0: return 1
    # memoized?
    key = a*25 + b
    if key in m: return m[key]
    # we already know that both a and b must be > 1
    n = combo(a, b-1) + combo(a-1, b)
    # memoize and return
    m[key] = n
    return n
        
a = int(raw_input().strip())
b = int(raw_input().strip())
print volleyball(a,b)
​
​
# solution 3
from operator import mul    # or mul=lambda x,y:x*y
from fractions import Fraction
​
def nCk(n,k): 
    return int( reduce(mul, (Fraction(n-i, i+1) for i in range(k)), 1) )
​
A=input()
B=input()
result = 0
winner = max(A,B)
loser = min(A,B)
​
if (winner == 25 and loser < 24):
    result = nCk(24+loser,loser)
elif winner > 25 and loser >= 24 and abs(A-B) == 2:
    result = nCk(48,24)
    if (A+B-48 > 0):
        result = result * pow(2,(A+B-50)/2,1000000007)
​
print result % 1000000007
​
​
# solution 4
C = {}
C[(0,0)] = 1
for n in range (1, 51):
    C[(n, 0)] = 1
    C[(n, n)] = 1
    for k in range (1, n):
        C[(n, k)] = (C[(n - 1, k - 1)] + C[(n - 1, k)])
​
def main():
    A = int(raw_input())
    B = int(raw_input())
    M = 10**9 + 7
    low = min(A, B)
    high = max(A, B)
​
    if low <= 23 and high == 25:
        if low > 0:
            print (C[(high + low, high)] - C[(high + low - 1, low - 1)]) % M
        else:
            print 1
    elif low > 23 and high == low + 2:
        print (C[(48,24)] * pow(2,low - 24,M)) % M
    else:
        print 0
​
if __name__ == "__main__":  
    main()
​
​
# solution 5
import operator as op
def ncr(n, r):
    r = min(r, n-r)
    if r == 0: return 1
    numer = reduce(op.mul, xrange(n, n-r, -1))
    denom = reduce(op.mul, xrange(1, r+1))
    return numer//denom
​
def solve(A, B):
    if B < 25:
        return 0
    if B == 25:
        if A < 24:
            return ncr(A + B - 1, A)
    if B - A == 2:
        return 32247603683100 * pow(2, A - 24, 10 ** 9 + 7)
    return 0
​
A, B = sorted([input(), input()])
print solve(A, B) % (10 ** 9 + 7)
​
​
​
# solution 6
import math
p=1000000007
​
fact=[1]
for x in xrange(1000000):
    fact.append( (x+1)*( fact[-1] ) %p)
​
    
def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b / a) * y, y)
​
def modinv(a, m):
    g, x, y = egcd(a, m)
    return x % m
​
​
def nCr(n,r):
    return (fact[n] * modinv((fact[r]*fact[n-r])%p ,p) ) %p
​
​
​
A=input()
B=input()
​
if max(A,B)<25 or abs(A-B)<2 or (min(A,B)>=24 and abs(A-B)!=2):
    print 0
elif min(A,B)<24:
    print nCr(A+B-1,min(A,B)) % p
else:
    temp=nCr(48,24)% p
    temp= (temp * pow(2,min(A,B)-24,p))%p
    print temp % p
​
​
​
# solution 7
def ball(x,y):
    a=max(x,y)
    b=min(x,y)
    import math
    if a<=24 and b<=24:
        return 0
    elif a==25 and b<24:
        c=math.factorial(a+b-1)/(math.factorial(a-1)*math.factorial(b))%1000000007
        return c
    elif a>=24 and b>=24:
        if a-b==2:
            d=math.factorial(48)/(math.factorial(24)*math.factorial(24))
            return (d*(2**(b-24)))%1000000007
        else:
            return 0
x=input()
y=input()
print ball(x,y)
Add Comment
```
